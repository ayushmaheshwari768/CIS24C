```python
commonWords = open("20k.txt")
mostUsedWords = commonWords.read().lower().split()
```

```python
def spellcheck(userWord):
    
    try:
        coord = {"q":{"y":0,'x':0},
                 "w":{"y":0,'x':1},
                 "e":{"y":0,'x':2},
                 "r":{"y":0,'x':3},
                 "t":{"y":0,'x':4},
                 "y":{"y":0,'x':5},
                 "u":{"y":0,'x':6},
                 "i":{"y":0,'x':7},
                 "o":{"y":0,'x':8},
                 "p":{"y":0,'x':9},
                 "a":{"y":1,'x':0},
                 "s":{"y":1,'x':1},
                 "d":{"y":1,'x':2},
                 "f":{"y":1,'x':3},
                 "g":{"y":1,'x':4},
                 "h":{"y":1,'x':5},
                 "j":{"y":1,'x':6},
                 "k":{"y":1,'x':7},
                 "l":{"y":1,'x':8},
                 "z":{"y":2,'x':0},
                 "x":{"y":2,'x':1},
                 "c":{"y":2,'x':2},
                 "v":{"y":2,'x':3},
                 "b":{"y":2,'x':4},
                 "n":{"y":2,'x':5},
                 "m":{"y":2,'x':6},}

        dictionary = {}
        k = 1
        for key in mostUsedWords:
            dictionary[key] = k
            k += 1

        if userWord == "" or userWord == None:
            return "Invalid Word"

        if userWord in mostUsedWords:
            return "No Error"
        
        else:
            minimumErr = None
            firstTime = True
            bestWords = []

            for word in mostUsedWords:

                error = 0
                addDel = 0
                replace = 0
                i = 0

                if len(word) >= len(userWord) - 1 and len(word) <= len(userWord) + 1:
                    
                    if len(word) == len(userWord) - 1 or len(word) == len(userWord) + 1:
                        addDel = 1
                                            
                    if word in userWord:
                        userWord = word
                    
                    for letter in userWord:
                        try:
                            if word[i] != letter:
                                if abs(coord[word[i]]["x"] - coord[letter]["x"]) + abs(coord[word[i]]["y"] - coord[letter]["y"]) <= 2:
                                    replace += 1
                                else:
                                    replace += 2
                            i += 1
                        except IndexError:
                            break

                    error = replace + addDel

                    if firstTime == True:
                        minimumErr = error
                        bestWords = [word]
                        firstTime = False
                    else:
                        if error == minimumErr:
                            bestWords.append(word)
                        elif error < minimumErr:
                            minimumErr = error
                            bestWords = [word]
            
            try:
                if minimumErr > 10:
                    return "Invalid Word"
                else:
                    topWord = ""

                    try:
                        if len(bestWords) > 1:
                            minIndex = 20002
                            for index in bestWords:
                                if dictionary[index] < minIndex:
                                    minIndex = dictionary[index]
                                    topWord = index
                        else:
                            topWord = bestWords[0]
                    except:
                        return "Invalid Word"

                    return ("Suggestion: " + topWord)
            except:
                return "Invalid Word"
        
    except KeyError:
        
        return "Invalid Word"
```

```python
userWord = input("Enter a word: ").lower()
print(spellcheck(userWord))
```
